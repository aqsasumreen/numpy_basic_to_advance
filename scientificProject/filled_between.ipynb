```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,3,1,4,5])
plt.plot(x,y)
# plt.fill_between(x,y)
# diff values of x,y se wo particular area me fill kry ga..
#  another way of putting  values is --> where  ,

# plt.fill_between(x =[2,4], y1 = 3, y2 = 5 , color ='red')
plt.fill_between(x , y,  color ='red', where=(x>= 2) & (x <= 4))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Language")
#  to save graph externally
# dpi means dots per inches , transparent = true se backgroud me b green color aye ga, 
#  we can save graph in png, img, pdf files
plt.savefig("Llin" , dpi = 2000, facecolor = "green", transparent = False, bbox_inches = "tight")
plt.show()
```
